```python
import numpy as np 
import matplotlib.pyplot  as plt
```

# hücre 2 veri hazırlığı 

```python
aylar=["Ocak"," Şubat", "Mart", "Nisan", "Mayıs"," Haziran", "Temmuz"," Ağustos", "Eylül", "Ekim"," Kasım"," Aralık"]
aylıksatıs=np.random.randint(200,15000,12)
urun=["kalem","bilgisayar","mouse","ekran kartı","monitör","saat","anahtar"]
urunsatıs=np.random.randint(100,500,7)
bolge=["Marmara","Ege","Akdeniz","İç Anadolu","Karadeniz"]
bolgesatıs=np.linspace(1021,5651,5)# linspace ile eşit aralıklarla yaptık
```

# hücre 3 lineplot


```python
plt.figure(figsize=(10,6)) # en  başta tanımlamamız lazım 
plt.plot(aylar,aylıksatıs,"r")
plt.xlabel("aylar")
plt.ylabel("aylık satışlar") 
plt.show()
```

# hücre 4 barplot

```python
plt.bar(urun,urunsatıs,color="#000080",)
plt.xlabel("ÜRÜNLER",color="blue",fontweight="bold")# kalın yazı demek bold 
plt.ylabel("ÜRÜN SATIŞLARI",color="red",fontsize=14) # sizeda ayarlarız
```

# hücre 5 grafik özelleştirme

```python
plt.plot(bolge,bolgesatıs,color="r",linestyle="--",marker="*",label="bölge bazlı satışlar")
plt.xlabel("bölgeler",fontsize=14)
plt.ylabel("bölge satışları",color="red",fontweight="bold",fontsize=14)
plt.legend(fontsize=10) # legend kalın yazı parametresini kabul etmezmiş 
```

```python
figure,axes=plt.subplots(nrows=2,ncols=2,figsize=(20,12))


ax1=axes[0,0]
ax1.plot(aylar,aylıksatıs,"r")
ax1.set_xlabel("aylar",fontweight="bold",color="blue")
ax1.set_ylabel("aylık satışlar",fontweight="bold",color="red")
ax1.set_title("aylık satış oranları ",fontsize=14)


ax2=axes[0,1] # satır 1 sütun 2 gibi düşğn cano
ax2.bar(urun,urunsatıs,width=0.5,color="green")# width değeri çizgilerin kalınlığını ayarlar
ax2.set_xlabel("ürünler",fontweight="bold")
ax2.set_ylabel("ürünsatışları",fontsize=15)
ax2.set_title("ürün satış oranları",fontweight="bold")
ax2.set_ylim(0, max(urunsatıs) + 200)  # biraz boşluk bırak # ürün satış değerindeki değeri fazlalaştırır
ax2.tick_params(axis='x', labelrotation=45) # etiket değerlerini 45 derece döndürdük


ax3=axes[1,0]
ax3.plot(bolge,bolgesatıs,"b")
ax3.set_xlabel("bölgeler",fontweight="bold",fontsize=(15))
ax3.set_ylabel("bölge satışları",fontsize=15)
ax3.set_title("bölge bazlı satış oranları",fontweight="bold")


#ax4=[1,1]
toplamsatıslar=[sum(aylıksatıs),sum(urunsatıs),sum(bolgesatıs)]
etıket=["aylıktop.","ürünlertop.","bölgelertop."]
ax4=axes[1,1]
ax4.bar(etıket,toplamsatıslar,width=0.5,color="blue")
ax4.set_xlabel("genel satışlar",fontweight="bold",fontsize=(15))
ax4.set_ylabel("toplam  satışları",fontsize=15)
ax4.set_title("genel özet",fontweight="bold",fontsize=(20))
ax4.set_ylim(0, max(toplamsatıslar) + 600)  # biraz boşluk bırak
```
